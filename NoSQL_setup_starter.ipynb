5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
def convert_to_float(document):
    if 'geocode' in document:
        if 'longitude' in document['geocode']:
            try:
                document['geocode']['longitude'] = float(Decimal(document['geocode']['longitude']))
            except (ValueError, InvalidOperation):
                pass  # Handle conversion error if necessary
        if 'latitude' in document['geocode']:
            try:
                document['geocode']['latitude'] = float(Decimal(document['geocode']['latitude']))
            except (ValueError, InvalidOperation):
                pass  # Handle conversion error if necessary
    return document
```

```python
# Retrieve documents and update the data type
for document in collection.find():
    updated_document = convert_to_float(document)
    
    # Update the document in the collection
    collection.update_one(
        {'_id': document['_id']},
        {'$set': updated_document}
    )

print("Update complete.")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
# Function to convert RatingValue fields to Integer
def convert_rating_value_to_int(document):
    if 'RatingValue' in document:
        try:
            # Check if RatingValue is a valid integer string
            if isinstance(document['RatingValue'], str):
                # Convert RatingValue from string to integer
                document['RatingValue'] = int(document['RatingValue'])
        except ValueError:
            # If conversion fails, set RatingValue to None
            document['RatingValue'] = None
    return document
```

```python
# Retrieve documents and update the data type
for document in collection.find():
    updated_document = convert_rating_value_to_int(document)
    
    # Update the document in the collection
    collection.update_one(
        {'_id': document['_id']},
        {'$set': updated_document}
    )

print("Update complete.")
```

```python
# Check that the coordinates and rating value are now numbers
# Function to print the type of fields
def print_field_types(document):
    if document:
        print("Data types of fields in the document:")
        if 'geocode' in document:
            if 'longitude' in document['geocode']:
                print(f"Longitude type: {type(document['geocode']['longitude'])}")
            if 'latitude' in document['geocode']:
                print(f"Latitude type: {type(document['geocode']['latitude'])}")
        if 'RatingValue' in document:
            print(f"RatingValue type: {type(document['RatingValue'])}")
    else:
        print("No document found.")

# Print the data types of the fields
print_field_types(document)
```

```python
# Retrieve documents and print types again
for document in collection.find({'RatingValue': {'$exists': True}}).limit(1):
    print_field_types(document)
```
